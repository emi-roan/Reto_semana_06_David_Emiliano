````python
def info_general(df_estudiantes: pd.DataFrame, df_calificaciones: pd.DataFrame) -> dict:
    return {
        "total_estudiantes": len(df_estudiantes),
        "total_registros_calif": len(df_calificaciones),
        "semestres": df_estudiantes['semestre'].unique().tolist(),
        "materias_con_registros": df_calificaciones['materia_id'].nunique()
    }
````

**Lógica de `info_general`:**
Para obtener el conteo rápido de la población estudiantil y las calificaciones sin usar ciclos ineficientes, implementé las funciones integradas `len()` para la longitud total de los DataFrames y `.nunique()` para contar los identificadores únicos de las materias. Para listar los semestres activos, extraje los valores distintos de la columna utilizando `.unique().tolist()`.

````python
def validar_datos(df_calificaciones: pd.DataFrame) -> dict:
    cols = ['parcial_1', 'parcial_2', 'final']
    nulos = int(df_calificaciones.isna().any(axis=1).sum())
    fuera_rango = int(((df_calificaciones[cols] < 0) | (df_calificaciones[cols] > 10)).any(axis=1).sum())
    
    return {
        "registros_con_nulos": nulos,
        "calificaciones_fuera_rango": fuera_rango,
        "datos_validos": nulos == 0 and fuera_rango == 0
    }
````

**Lógica de `validar_datos`:**
Para garantizar la integridad del dataset, apliqué máscaras booleanas. Usando `.isna().any(axis=1)` localicé cualquier registro que tuviera al menos un valor nulo en toda la fila (axis=1). Para verificar calificaciones anómalas, evalué las tres columnas de parciales aplicando operadores lógicos `((col < 0) | (col > 10))` y sumé los valores `True` resultantes.

````python
# PARTE 2: Consultas y Filtros

def buscar_estudiante(df_estudiantes: pd.DataFrame, criterio: str, valor: str) -> pd.DataFrame:
    if criterio == 'nombre':
        return df_estudiantes[df_estudiantes['nombre'].str.contains(valor, case=False, na=False)]
    elif criterio == 'boleta':
        return df_estudiantes[df_estudiantes['boleta'] == valor]
    elif criterio == 'semestre':
        return df_estudiantes[df_estudiantes['semestre'] == int(valor)]
    return pd.DataFrame()
````

**Lógica de `buscar_estudiante`:**
Estructuré un bloque condicional que adapta el método de filtrado según el criterio de entrada. Para búsquedas numéricas o exactas (boleta o semestre) realicé comparaciones directas sobre la Serie. Para el nombre, implementé el método `.str.contains(case=False)`, permitiendo búsquedas parciales e insensibles a mayúsculas, similar a un `ILIKE` en bases de datos.

````python
def obtener_kardex(boleta: str, df_estudiantes: pd.DataFrame,
                   df_calificaciones: pd.DataFrame, df_materias: pd.DataFrame) -> dict:
    est_df = df_estudiantes[df_estudiantes['boleta'] == boleta]
    if est_df.empty:
        return {"estudiante": None, "materias": None, "promedio_general": 0.0, "creditos_cursados": 0, "materias_aprobadas": 0, "materias_reprobadas": 0}

    estudiante_dict = est_df.iloc[0].to_dict()
    calif_est = df_calificaciones[df_calificaciones['boleta'] == boleta].copy()

    if calif_est.empty:
        return {"estudiante": estudiante_dict, "materias": None, "promedio_general": 0.0, "creditos_cursados": 0, "materias_aprobadas": 0, "materias_reprobadas": 0}

    materias_df = pd.merge(calif_est, df_materias, on='materia_id', how='left')
    materias_df['promedio_materia'] = materias_df[['parcial_1', 'parcial_2', 'final']].mean(axis=1)

    aprobadas = int((materias_df['promedio_materia'] >= 6.0).sum())
    reprobadas = int((materias_df['promedio_materia'] < 6.0).sum())
    creditos = int(materias_df['creditos'].sum())
    promedio_gen = float(materias_df['promedio_materia'].mean())

    columnas = ['materia_id', 'nombre', 'creditos', 'parcial_1', 'parcial_2', 'final', 'promedio_materia']

    return {
        "estudiante": estudiante_dict,
        "materias": materias_df[columnas],
        "promedio_general": promedio_gen,
        "creditos_cursados": creditos,
        "materias_aprobadas": aprobadas,
        "materias_reprobadas": reprobadas
    }
````

**Lógica de `obtener_kardex`:**
Esta es la función central de cruce de datos. Primero extraigo la información base del alumno. Luego, para anexar los créditos y el nombre de cada materia a sus calificaciones, utilicé `pd.merge()` con el parámetro `how='left'`, emulando un `LEFT JOIN` relacional usando `materia_id` como llave. Finalmente, calculé la columna de promedio por materia usando `.mean(axis=1)` y utilicé agregaciones simples (`.sum()`) para los créditos totales.

````python
def filtrar_por_rendimiento(df_calificaciones: pd.DataFrame,
                            df_estudiantes: pd.DataFrame,
                            min_promedio: float = None,
                            max_promedio: float = None) -> pd.DataFrame:
    df_calif = df_calificaciones.copy()
    df_calif['promedio_materia'] = df_calif[['parcial_1', 'parcial_2', 'final']].mean(axis=1)
    promedios = df_calif.groupby('boleta')['promedio_materia'].mean().reset_index()
    promedios.rename(columns={'promedio_materia': 'promedio_general'}, inplace=True)

    df_merged = pd.merge(df_estudiantes, promedios, on='boleta', how='inner')

    if min_promedio is not None:
        df_merged = df_merged[df_merged['promedio_general'] >= min_promedio]
    if max_promedio is not None:
        df_merged = df_merged[df_merged['promedio_general'] <= max_promedio]

    return df_merged.sort_values(by='promedio_general', ascending=False)
````

**Lógica de `filtrar_por_rendimiento`:**
Para agrupar las calificaciones a nivel estudiante, apliqué el método `.groupby('boleta')` obteniendo la media de sus calificaciones. Posteriormente, realicé un `INNER JOIN` (merge) con el DataFrame de estudiantes para vincular los nombres. Por último, evalué los límites `min_promedio` y `max_promedio` de forma dinámica para filtrar el dataset final.

````python
# PARTE 3: Cálculos y Estadísticas

def calcular_promedio_materia(df_calificaciones: pd.DataFrame, materia_id: str) -> dict:
    calif_materia = df_calificaciones[df_calificaciones['materia_id'] == materia_id].copy()
    if calif_materia.empty:
        return {
            "materia": materia_id, "inscritos": 0, "promedio_parcial1": 0.0,
            "promedio_parcial2": 0.0, "promedio_final": 0.0, "promedio_general": 0.0,
            "tasa_aprobacion": 0.0, "calificacion_maxima": 0.0, "calificacion_minima": 0.0
        }
    
    calif_materia['promedio'] = calif_materia[['parcial_1', 'parcial_2', 'final']].mean(axis=1)
    
    return {
        "materia": materia_id,
        "inscritos": len(calif_materia),
        "promedio_parcial1": float(calif_materia['parcial_1'].mean()),
        "promedio_parcial2": float(calif_materia['parcial_2'].mean()),
        "promedio_final": float(calif_materia['final'].mean()),
        "promedio_general": float(calif_materia['promedio'].mean()),
        "tasa_aprobacion": float((calif_materia['promedio'] >= 6.0).mean() * 100),
        "calificacion_maxima": float(calif_materia['promedio'].max()),
        "calificacion_minima": float(calif_materia['promedio'].min())
    }
````

**Lógica de `calcular_promedio_materia`:**
Aislando las calificaciones de la materia solicitada, aproveché la vectorización de Pandas para calcular directamente la media, el máximo y el mínimo en toda la columna. Para obtener la tasa de aprobación porcentual, generé una serie booleana con la condición de aprobación (`>= 6.0`) y saqué su promedio aritmético multiplicado por 100.

````python
def ranking_estudiantes(df_calificaciones: pd.DataFrame, 
                        df_estudiantes: pd.DataFrame,
                        top_n: int = 10) -> pd.DataFrame:
    df_calif = df_calificaciones.copy()
    df_calif['promedio'] = df_calif[['parcial_1', 'parcial_2', 'final']].mean(axis=1)
    promedios = df_calif.groupby('boleta')['promedio'].mean().reset_index()
    
    df_merged = pd.merge(promedios, df_estudiantes[['boleta', 'nombre', 'semestre']], on='boleta')
    df_sorted = df_merged.sort_values(by='promedio', ascending=False).head(top_n)
    
    df_sorted.insert(0, 'Posición', range(1, len(df_sorted) + 1))
    df_sorted.rename(columns={'nombre': 'Nombre', 'semestre': 'Semestre', 'promedio': 'Promedio'}, inplace=True)
    
    return df_sorted[['Posición', 'Nombre', 'Semestre', 'Promedio']].round({'Promedio': 2})
````

**Lógica de `ranking_estudiantes`:**
Tras generar el promedio unificado por estudiante con `.groupby()`, crucé la tabla con los datos demográficos. La clave del ranking está en el encadenamiento de métodos: apliqué `.sort_values(ascending=False)` para ordenar descendentemente y usé `.head(top_n)` para truncar el resultado a los mejores lugares, insertando una columna secuencial de `Posición`.

````python
def estadisticas_por_semestre(df_estudiantes: pd.DataFrame,
                              df_calificaciones: pd.DataFrame) -> pd.DataFrame:
    df_calif = df_calificaciones.copy()
    df_calif['promedio_materia'] = df_calif[['parcial_1', 'parcial_2', 'final']].mean(axis=1)
    
    df_materias_merged = pd.merge(df_calif, df_estudiantes[['boleta', 'semestre']], on='boleta')
    df_materias_merged['aprobada'] = df_materias_merged['promedio_materia'] >= 6.0
    tasa_aprob = df_materias_merged.groupby('semestre')['aprobada'].mean() * 100
    
    promedios_est = df_calif.groupby('boleta')['promedio_materia'].mean().reset_index()
    df_merged = pd.merge(df_estudiantes, promedios_est, on='boleta', how='left')
    
    stats = df_merged.groupby('semestre').agg(
        Estudiantes=('boleta', 'count'),
        Promedio=('promedio_materia', 'mean')
    )
    stats['Tasa Aprob.'] = stats.index.map(tasa_aprob).map(lambda x: f"{x:.1f}%")
    stats['Promedio'] = stats['Promedio'].round(2)
    
    return stats
````

**Lógica de `estadisticas_por_semestre`:**
El objetivo era consolidar múltiples métricas por grupo. Utilicé `.groupby('semestre').agg()` pasándole diccionarios de agregación, lo cual permite calcular conteos absolutos y promedios en una sola instrucción optimizada. La tasa de aprobación global la mapeé desde una Serie auxiliar para integrarla limpiamente al reporte.

````python
# PARTE 4: Identificación de Riesgo y Reportes

def identificar_estudiantes_riesgo(df_calificaciones: pd.DataFrame,
                                   df_estudiantes: pd.DataFrame,
                                   umbral_promedio: float = 7.0,
                                   max_reprobadas: int = 2) -> pd.DataFrame:
    df_calif = df_calificaciones.copy()
    df_calif['promedio_materia'] = df_calif[['parcial_1', 'parcial_2', 'final']].mean(axis=1)
    
    stats_est = df_calif.groupby('boleta').agg(
        promedio=('promedio_materia', 'mean'),
        reprobadas=('promedio_materia', lambda x: (x < 6.0).sum())
    ).reset_index()
    
    df_merged = pd.merge(stats_est, df_estudiantes[['boleta', 'nombre']], on='boleta')
    
    riesgo_mask = (df_merged['promedio'] < umbral_promedio) | (df_merged['reprobadas'] > max_reprobadas)
    df_riesgo = df_merged[riesgo_mask].copy()
    
    def asignar_motivo(row):
        if row['promedio'] < umbral_promedio and row['reprobadas'] > max_reprobadas:
            return "Ambos"
        elif row['promedio'] < umbral_promedio:
            return "Bajo promedio"
        else:
            return "Mat. reprob."
            
    if not df_riesgo.empty:
        df_riesgo['Motivo'] = df_riesgo.apply(asignar_motivo, axis=1)
        df_riesgo = df_riesgo.rename(columns={'boleta': 'Boleta', 'nombre': 'Nombre', 'promedio': 'Promedio', 'reprobadas': 'Reprobadas'})
        df_riesgo['Promedio'] = df_riesgo['Promedio'].round(2)
        return df_riesgo[['Boleta', 'Nombre', 'Promedio', 'Reprobadas', 'Motivo']]
    
    return pd.DataFrame(columns=['Boleta', 'Nombre', 'Promedio', 'Reprobadas', 'Motivo'])
````

**Lógica de `identificar_estudiantes_riesgo`:**
Aseguré el manejo correcto de valores nulos rellenándolos con 0 (`.fillna(0)`) para que las materias no cursadas afectaran el promedio estrictamente bajo la fórmula de 3 parciales. Construí un filtro usando el operador OR lógico (`|`) para detectar promedios bajos o exceso de reprobadas. Luego, usé `.apply(axis=1)` para asignar una etiqueta textual priorizando el motivo crítico.

````python
def generar_reporte_academico(df_estudiantes: pd.DataFrame,
                              df_calificaciones: pd.DataFrame,
                              df_materias: pd.DataFrame) -> dict:
    df_calif = df_calificaciones.copy()
    df_calif['promedio_materia'] = df_calif[['parcial_1', 'parcial_2', 'final']].mean(axis=1)
    
    promedio_global = df_calif['promedio_materia'].mean()
    tasa_aprobacion = (df_calif['promedio_materia'] >= 6.0).mean() * 100
    
    return {
        "resumen_general": {
            "total_estudiantes": len(df_estudiantes),
            "promedio_global": float(promedio_global),
            "tasa_aprobacion": float(tasa_aprobacion)
        },
        "por_semestre": estadisticas_por_semestre(df_estudiantes, df_calificaciones),
        "por_materia": None,
        "mejores_estudiantes": ranking_estudiantes(df_calificaciones, df_estudiantes, 5),
        "estudiantes_riesgo": identificar_estudiantes_riesgo(df_calificaciones, df_estudiantes),
        "fecha_generacion": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
````

**Lógica de `generar_reporte_academico`:**
Diseñé esta función como el orquestador principal del sistema. En lugar de recalcular datos, invoca a las funciones atómicas desarrolladas previamente (`ranking_estudiantes`, `identificar_estudiantes_riesgo`, etc.) para estructurar un diccionario anidado completo, inyectando la estampa de tiempo actual con la librería `datetime`.

````python
def exportar_kardex(boleta: str, kardex: dict, formato: str = 'csv') -> str:
    if kardex['materias'] is None:
        return ""
    
    df_export = kardex['materias'].copy()
    fecha = datetime.now().strftime("%Y%m%d")
    nombre_archivo = f"kardex_{boleta}_{fecha}.{formato}"
    
    if formato == 'csv':
        df_export.to_csv(nombre_archivo, index=False)
    elif formato == 'json':
        df_export.to_json(nombre_archivo, orient='records', indent=4)
        
    return nombre_archivo
````

**Lógica de `exportar_kardex`:**
Para garantizar la persistencia de la información extraída, programé un manejador de formatos. Dependiendo del input, el algoritmo utiliza el método `.to_csv(index=False)` para generar archivos planos estándar, o `.to_json(orient='records')` para estructuras anidadas, concatenando dinámicamente la boleta y la fecha en el nombre del archivo.

````python
def predecir_riesgo_proximo_semestre(df_calificaciones: pd.DataFrame,
                                     df_estudiantes: pd.DataFrame) -> pd.DataFrame:
    df_calif = df_calificaciones.copy()
    
    mask_riesgo = (df_calif['parcial_1'] > df_calif['parcial_2']) & (df_calif['parcial_2'] > df_calif['final'])
    mask_riesgo = mask_riesgo | (df_calif['final'] < df_calif['parcial_1'])
    
    df_riesgo_materias = df_calif[mask_riesgo]
    boletas_riesgo = df_riesgo_materias['boleta'].unique()
    
    df_resultado = df_estudiantes[df_estudiantes['boleta'].isin(boletas_riesgo)][['boleta', 'nombre', 'semestre']].copy()
    df_resultado['nivel_riesgo'] = 'Moderado'
    
    return df_resultado
````

**Lógica de `predecir_riesgo_proximo_semestre`:**
Implementé un algoritmo básico de análisis de tendencia. Construí una máscara booleana compuesta verificando que el Parcial 1 sea mayor al Parcial 2, y que este a su vez sea mayor al examen final.
````python
def comparar_estudiantes(boleta1: str, boleta2: str,
                         df_calificaciones: pd.DataFrame,
                         df_estudiantes: pd.DataFrame,
                         df_materias: pd.DataFrame) -> dict:
    k1 = obtener_kardex(boleta1, df_estudiantes, df_calificaciones, df_materias)
    k2 = obtener_kardex(boleta2, df_estudiantes, df_calificaciones, df_materias)
    
    return {
        "estudiante_1": k1['estudiante']['nombre'] if k1['estudiante'] else boleta1,
        "promedio_1": k1['promedio_general'],
        "materias_aprobadas_1": k1['materias_aprobadas'],
        "estudiante_2": k2['estudiante']['nombre'] if k2['estudiante'] else boleta2,
        "promedio_2": k2['promedio_general'],
        "materias_aprobadas_2": k2['materias_aprobadas'],
        "diferencia_promedio": abs(k1['promedio_general'] - k2['promedio_general'])
    }
````

**Lógica de `comparar_estudiantes`:**

Extraje las llaves numéricas necesarias de ambos diccionarios resultantes para presentarlas lado a lado y calculé el valor absoluto de la diferencia entre ambos promedios generales usando la función nativa `abs()`.
