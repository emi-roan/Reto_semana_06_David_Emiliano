```python
    """
    Analiza los rendimientos calculados.
    
    Retorna:
        {

            "volatilidad": float             # Desv. std de rendimientos
        }
    """
```

**Explicación de la lógica (Parte 1):**
* `estadisticas_basicas`: Se emplearon métodos vectorizados nativos de Pandas (`.min()`, `.max()`, `.mean()`, etc.) para calcular las medidas estadísticas directamente sobre la Serie, optimizando el tiempo de ejecución.
* `calcular_rendimientos`: Se aplicó `.pct_change()` para obtener la variación fraccional entre periodos consecutivos, multiplicando por 100 para estandarizar el formato a porcentaje.
* `analisis_rendimientos`: Se utilizaron métodos de agregación y máscaras booleanas (`rendimientos > 0`) para filtrar y cuantificar los días de ganancias y pérdidas, extrayendo las fechas extremas con `.idxmax()` e `.idxmin()`.

```python
def media_movil(precios: pd.Series, ventana: int) -> pd.Series:
    """
    Calcula la media móvil simple (SMA).
    
    Args:
        precios: Serie de precios
        ventana: Número de períodos para el promedio
    
    Retorna: Serie con la media móvil
    """

def bandas_bollinger(precios: pd.Series, ventana: int = 20, num_std: int = 2) -> dict:
    """
    Calcula las Bandas de Bollinger.
    
    - Banda media: SMA de 'ventana' períodos
    - Banda superior: SMA + (num_std × desviación estándar)
    - Banda inferior: SMA - (num_std × desviación estándar)
    
    Retorna:
        {
            "banda_superior": pd.Series,
            "banda_media": pd.Series,
            "banda_inferior": pd.Series
        }
    """

def detectar_maximos_minimos(precios: pd.Series, ventana: int = 5) -> dict:
    """
    Detecta máximos y mínimos locales.
    
    Un máximo local: precio mayor que los 'ventana' días anteriores y posteriores
    Un mínimo local: precio menor que los 'ventana' días anteriores y posteriores
    
    Retorna:
        {
            "maximos": pd.Series,  # Fechas y precios de máximos
            "minimos": pd.Series   # Fechas y precios de mínimos
        }
    """

def clasificar_tendencia(precios: pd.Series, ventana: int = 10) -> str:
    """
    Clasifica la tendencia actual.
    
    Criterios:
    - ALCISTA: Precio actual > MA y MA subiendo
    - BAJISTA: Precio actual < MA y MA bajando
    - LATERAL: Sin tendencia clara
    
    Retorna: "ALCISTA", "BAJISTA" o "LATERAL"
    """
```

**Explicación de la lógica (Parte 2):**
* `media_movil` y `bandas_bollinger`: Se implementó `.rolling(window).mean()` y `.std()` para generar ventanas móviles, permitiendo suavizar la curva de precios y calcular los límites de desviación estándar para las bandas.
* `detectar_maximos_minimos`: Se comparó la serie original contra una ventana móvil centrada (`center=True`) para encontrar los puntos exactos donde el precio actual coincide con el máximo o mínimo local de dicha ventana.
* `clasificar_tendencia`: Se evaluó la posición del precio de cierre más reciente frente a su media móvil, verificando adicionalmente la pendiente de la media (comparando el último valor con el penúltimo) para confirmar la dirección.

## Código Base

````python
import pandas as pd
import numpy as np
from typing import Dict, List, Tuple, Optional
````

````python
# PARTE 1: Análisis Estadístico Básico

def estadisticas_basicas(precios: pd.Series) -> dict:
    return {
        "precio_actual": float(precios.iloc[-1]),
        "precio_minimo": float(precios.min()),
        "precio_maximo": float(precios.max()),
        "precio_promedio": float(precios.mean()),
        "precio_mediana": float(precios.median()),
        "desviacion_std": float(precios.std()),
        "rango": float(precios.max() - precios.min()),
        "dias_analizados": len(precios)
    }
````

````python
def calcular_rendimientos(precios: pd.Series) -> pd.Series:
    return precios.pct_change() * 100
````

````python
def analisis_rendimientos(rendimientos: pd.Series) -> dict:
    return {
        "rendimiento_total": float(rendimientos.sum()),
        "rendimiento_promedio": float(rendimientos.mean()),
        "mejor_dia": (rendimientos.idxmax(), float(rendimientos.max())),
        "peor_dia": (rendimientos.idxmin(), float(rendimientos.min())),
        "dias_positivos": int((rendimientos > 0).sum()),
        "dias_negativos": int((rendimientos < 0).sum()),
        "volatilidad": float(rendimientos.std())
    }
    
    return resultado
````

**Explicación de la lógica (Parte 1):**
* `estadisticas_basicas`: Se emplearon métodos vectorizados nativos de Pandas (.min, .max, .mean, etc.) para calcular las medidas estadísticas directamente sobre la Serie, optimizando el tiempo de ejecución.
* `calcular_rendimientos`: Se aplicó .pct_change() para obtener la variación fraccional entre periodos consecutivos, multiplicando por 100 para estandarizar el formato a porcentaje.
* `analisis_rendimientos`: Se utilizaron métodos de agregación y máscaras booleanas (rendimientos > 0) para filtrar y cuantificar los días de ganancias y pérdidas, extrayendo las fechas extremas con .idxmax() e .idxmin().

````python
# PARTE 2: Indicadores Técnicos

def media_movil(precios: pd.Series, ventana: int) -> pd.Series:
    return precios.rolling(window=ventana).mean()
````

````python
def bandas_bollinger(precios: pd.Series, ventana: int = 20, num_std: int = 2) -> dict:
    sma = precios.rolling(window=ventana).mean()
    std = precios.rolling(window=ventana).std()
    return {
        "banda_superior": sma + (num_std * std),
        "banda_media": sma,
        "banda_inferior": sma - (num_std * std)
    }
    return resultado
````

````python
def detectar_maximos_minimos(precios: pd.Series, ventana: int = 5) -> dict:
    max_locales = precios[precios == precios.rolling(window=ventana * 2 + 1, center=True).max()]
    min_locales = precios[precios == precios.rolling(window=ventana * 2 + 1, center=True).min()]
    return {
        "maximos": max_locales,
        "minimos": min_locales
    }
    return resultado
````

````python
def clasificar_tendencia(precios: pd.Series, ventana: int = 10) -> str:
    ma = precios.rolling(window=ventana).mean()
    precio_actual = precios.iloc[-1]
    ma_actual = ma.iloc[-1]
    ma_anterior = ma.iloc[-2]
    
    if precio_actual > ma_actual and ma_actual > ma_anterior:
        return "ALCISTA"
    elif precio_actual < ma_actual and ma_actual < ma_anterior:
        return "BAJISTA"
    else:
        return "LATERAL"
````

**Explicación de la lógica (Parte 2):**
* `media_movil` y `bandas_bollinger`: Se implementó .rolling(window).mean() y .std() para generar ventanas móviles, permitiendo suavizar la curva de precios y calcular los límites de desviación estándar para las bandas.
* `detectar_maximos_minimos`: Se comparó la serie original contra una ventana móvil centrada (center=True) para encontrar los puntos exactos donde el precio actual coincide con el máximo o mínimo local de dicha ventana.
* `clasificar_tendencia`: Se evaluó la posición del precio de cierre más reciente frente a su media móvil, verificando adicionalmente la pendiente de la media (comparando el último valor con el penúltimo) para confirmar la dirección.

````python
# PARTE 3: Sistema de Alertas

def generar_senales_trading(precios: pd.Series, ma_corta: int = 5, ma_larga: int = 20) -> pd.Series:
    ma_c = precios.rolling(window=ma_corta).mean()
    ma_l = precios.rolling(window=ma_larga).mean()
    
    # Empezamos asumiendo que todo el tiempo la señal es "MANTENER"
    senales = pd.Series("MANTENER", index=precios.index)
    
    # Detectamos los cruces usando shift(1) para ver el valor del día anterior
    cruce_arriba = (ma_c > ma_l) & (ma_c.shift(1) <= ma_l.shift(1))
    cruce_abajo = (ma_c < ma_l) & (ma_c.shift(1) >= ma_l.shift(1))
    
    senales[cruce_arriba] = "COMPRA"
    senales[cruce_abajo] = "VENTA"
    
    return senales
````

````python
def alertas_precio(precios: pd.Series, umbral_cambio: float = 5.0) -> list:
    rendimientos = calcular_rendimientos(precios)
    alertas = []
    
    # Iteramos sobre los rendimientos buscando los que superen el umbral
    for fecha, rend in rendimientos.items():
        if pd.notna(rend) and abs(rend) >= umbral_cambio:
            tipo = "SUBIDA" if rend > 0 else "CAIDA"
            fecha_str = fecha.strftime('%Y-%m-%d') if hasattr(fecha, 'strftime') else str(fecha)
            alertas.append({
                "fecha": fecha_str,
                "tipo": tipo,
                "cambio": float(rend)
            })
            
    return alertas
````

````python
def clasificar_volatilidad(rendimientos: pd.Series) -> str:
    vol = rendimientos.std()
    
    if pd.isna(vol):
        return "DESCONOCIDA"
    if vol < 1.0:
        return "BAJA"
    elif vol <= 3.0:
        return "MEDIA"
    elif vol <= 5.0:
        return "ALTA"
    else:
        return "MUY ALTA"
````

````python
def generar_reporte_completo(precios: pd.Series, nombre_accion: str) -> dict:
    # Usamos todas las funciones anteriores para armar un súper diccionario
    rendimientos = calcular_rendimientos(precios)
    fecha_inicio = precios.index[0]
    fecha_fin = precios.index[-1]
    
    # Formateamos las fechas a texto
    inicio_str = fecha_inicio.strftime('%Y-%m-%d') if hasattr(fecha_inicio, 'strftime') else str(fecha_inicio)
    fin_str = fecha_fin.strftime('%Y-%m-%d') if hasattr(fecha_fin, 'strftime') else str(fecha_fin)
    
    return {
        "nombre": nombre_accion,
        "periodo": {
            "inicio": inicio_str,
            "fin": fin_str,
            "dias": len(precios)
        },
        "estadisticas": estadisticas_basicas(precios),
        "rendimientos": analisis_rendimientos(rendimientos),
        "tendencia": clasificar_tendencia(precios),
        "volatilidad": clasificar_volatilidad(rendimientos),
        "senal_actual": generar_senales_trading(precios).iloc[-1],
        "alertas_recientes": alertas_precio(precios)
    }
````

**Explicación de la lógica (Parte 3):**
* `generar_senales_trading`: Se compararon dos medias móviles (rápida y lenta). Se utilizó .shift(1) para evaluar el estado del día anterior y así poder detectar el momento exacto del cruce, asignando las etiquetas de COMPRA o VENTA.
* `alertas_precio`: Se iteró sobre la serie de rendimientos previamente calculada, filtrando únicamente aquellos valores absolutos que superan el umbral definido para estructurarlos en una lista de diccionarios.
* `clasificar_volatilidad` y `generar_reporte`: Se consolidaron todas las funciones modulares. La volatilidad se segmentó mediante rangos condicionales sobre la desviación estándar, y el reporte integra todos los diccionarios y series en una sola estructura de datos legible.

## Funciones de Visualización

````python
def mostrar_reporte(reporte: Dict) -> None:
    """Muestra el reporte de forma legible."""
    print("=" * 70)
    print(f"           REPORTE DE ANÁLISIS: {reporte['nombre']}")
    print("=" * 70)
    
    # Período
    periodo = reporte.get('periodo', {})
    print(f"\n📅 PERÍODO DE ANÁLISIS")
    print("-" * 40)
    print(f"Inicio: {periodo.get('inicio', 'N/A')}")
    print(f"Fin: {periodo.get('fin', 'N/A')}")
    print(f"Días analizados: {periodo.get('dias', 'N/A')}")
    
    # Estadísticas
    stats = reporte.get('estadisticas', {})
    print(f"\n📊 ESTADÍSTICAS DE PRECIO")
    print("-" * 40)
    print(f"Precio actual:  ${stats.get('precio_actual', 0):,.2f}")
    print(f"Precio mínimo:  ${stats.get('precio_minimo', 0):,.2f}")
    print(f"Precio máximo:  ${stats.get('precio_maximo', 0):,.2f}")
    print(f"Precio promedio: ${stats.get('precio_promedio', 0):,.2f}")
    
    # Rendimientos
    rend = reporte.get('rendimientos', {})
    print(f"\n📈 RENDIMIENTO")
    print("-" * 40)
    print(f"Rendimiento total: {rend.get('rendimiento_total', 0):+.2f}%")
    print(f"Rendimiento promedio diario: {rend.get('rendimiento_promedio', 0):+.3f}%")
    if rend.get('mejor_dia'):
        print(f"Mejor día: {rend['mejor_dia'][0]} ({rend['mejor_dia'][1]:+.2f}%)")
    if rend.get('peor_dia'):
        print(f"Peor día: {rend['peor_dia'][0]} ({rend['peor_dia'][1]:+.2f}%)")
    print(f"Días positivos: {rend.get('dias_positivos', 0)}")
    print(f"Días negativos: {rend.get('dias_negativos', 0)}")
    
    # Indicadores
    print(f"\n🎯 INDICADORES")
    print("-" * 40)
    print(f"Tendencia: {reporte.get('tendencia', 'N/A')}")
    print(f"Volatilidad: {reporte.get('volatilidad', 'N/A')}")
    print(f"Señal actual: {reporte.get('senal_actual', 'N/A')}")
    
    # Alertas
    alertas = reporte.get('alertas_recientes', [])
    if alertas:
        print(f"\n⚠️ ALERTAS RECIENTES")
        print("-" * 40)
        for alerta in alertas[-5:]:  # Últimas 5
            emoji = "🔺" if alerta['tipo'] == 'SUBIDA' else "🔻"
            print(f"{emoji} {alerta['fecha']}: {alerta['tipo']} de {alerta['cambio']:+.2f}%")
    
    print("\n" + "=" * 70)
````

````python
def visualizar_precios_texto(precios: pd.Series, ancho: int = 50) -> None:
    """Visualización simple de precios en texto (ASCII chart)."""
    min_precio = precios.min()
    max_precio = precios.max()
    rango = max_precio - min_precio
    
    print(f"\nGráfico de precios: {precios.name}")
    print(f"Max: ${max_precio:.2f}")
    print("-" * (ancho + 10))
    
    # Mostrar cada 3 días para no saturar
    for fecha, precio in precios.iloc[::3].items():
        posicion = int((precio - min_precio) / rango * ancho) if rango > 0 else ancho // 2
        barra = " " * posicion + "█"
        fecha_str = fecha.strftime('%m/%d') if hasattr(fecha, 'strftime') else str(fecha)[:5]
        print(f"{fecha_str} |{barra}")
    
    print("-" * (ancho + 10))
    print(f"Min: ${min_precio:.2f}")
````

## Bonus: Funcionalidades Extra (Opcional)

### 1. RSI (Relative Strength Index)

````python
def calcular_rsi(precios: pd.Series, periodos: int = 14) -> pd.Series:
    delta = precios.diff()
    ganancias = delta.where(delta > 0, 0.0)
    perdidas = -delta.where(delta < 0, 0.0)
    
    promedio_ganancias = ganancias.rolling(window=periodos, min_periods=1).mean()
    promedio_perdidas = perdidas.rolling(window=periodos, min_periods=1).mean()
    
    rs = promedio_ganancias / promedio_perdidas
    rsi = 100 - (100 / (1 + rs))
    
    return rsi.fillna(50)
````

### 2. Backtesting simple

````python
def backtest_estrategia(precios: pd.Series, senales: pd.Series, capital_inicial: float = 10000) -> dict:
    capital = capital_inicial
    acciones = 0
    operaciones = 0
    ganadoras = 0
    precio_compra = 0.0

    for i in range(1, len(precios)):
        senal = senales.iloc[i]
        precio = float(precios.iloc[i])

        if senal == "COMPRA" and capital >= precio:
            acciones = capital // precio
            capital -= acciones * precio
            precio_compra = precio
            
        elif senal == "VENTA" and acciones > 0:
            capital += acciones * precio
            operaciones += 1
            if precio > precio_compra:
                ganadoras += 1
            acciones = 0

    capital_final = capital + (acciones * float(precios.iloc[-1]))
    rendimiento = ((capital_final - capital_inicial) / capital_inicial) * 100

    return {
        "capital_final": float(capital_final),
        "rendimiento_total": float(rendimiento),
        "num_operaciones": int(operaciones),
        "operaciones_ganadoras": int(ganadoras)
    }
````

**Explicación de la lógica (Bonus):**
* `calcular_rsi`: Se separaron las variaciones positivas y negativas del precio usando `.diff()` y `.where()`. Se calculó el promedio móvil de ambas para obtener la Fuerza Relativa (RS) y aplicar la fórmula estándar del oscilador de 0 a 100.
* `backtest_estrategia`: Se programó un motor iterativo que recorre las señales generadas, simulando compras con el capital disponible cuando hay señal alcista y liquidando posiciones cuando hay señal bajista, calculando finalmente la rentabilidad neta de la estrategia.
